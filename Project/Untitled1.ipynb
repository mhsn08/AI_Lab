```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('Dataset.txt', sep=",")
df.columns
```

```python
colors = {'normal': 'red', 'dos': 'blue', 'r2l': 'green', 'probe': 'orange', 'u2r': 'purple'}

fig, ax = plt.subplots()
for attack_type, group in df.groupby('attack_type'):
    ax.scatter(group['src_bytes'], group['dst_bytes'], c=colors[attack_type], label=attack_type)
ax.legend()
ax.set_xlabel('src_bytes')
ax.set_ylabel('dst_bytes')
plt.show()
```
