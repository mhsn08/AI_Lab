```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)

# generate 500 random x-values between 0 and 100
x = np.random.randint(0, 101, 500)

# generate corresponding y-values
y = x + np.random.normal(0, 10, 500)

# create scatter plot
plt.scatter(x, y)

# set title and labels
plt.title("Scatter Plot of random values")
plt.xlabel("X Values")
plt.ylabel("Y Values")

# set axis limits
plt.xlim(0, 100)
plt.ylim(0, 200)

# show plot
plt.show()
```

```python
sales_data = {
    'Product': ['TV','Phone','Shirt','Pants','Fruits','Vegetables'],
    'Revenue': [5000,3000,1000,2000,1500,2500]
}

sales_data = pd.DataFrame(sales_data)
sales_data.head()

plt.title('Total Revenue by product',color='blue')
plt.xlabel('Revenue',color='blue')
plt.ylabel('Product',color='blue')

plt.barh(sales_data['Product'],sales_data['Revenue'])
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# create example dataframe with three subjects and five students
data = {'Student': ['Alice', 'Bob', 'Charlie', 'Dave', 'Eve'],
        'Maths': [85, 90, 78, 92, 88],
        'English': [92, 85, 90, 87, 78],
        'Science': [89, 88, 92, 80, 85]}
df = pd.DataFrame(data)

# calculate average marks of each subject
avg_marks = df[['Maths', 'English', 'Science']].mean()

# create bar chart
plt.bar(avg_marks.index, avg_marks)

# set title and labels
plt.title('Average Marks by Subject')
plt.xlabel('Subject')
plt.ylabel('Average Marks')

# show plot
plt.show()
```

```python
import pandas as pd

# example list of dictionaries with student names and marks
students = [{'Name': 'Alice', 'Maths': 85, 'English': 92, 'Science': 89},
            {'Name': 'Bob', 'Maths': 90, 'English': 85, 'Science': 88},
            {'Name': 'Charlie', 'Maths': 78, 'English': 90, 'Science': 92},
            {'Name': 'Dave', 'Maths': 92, 'English': 87, 'Science': 80},
            {'Name': 'Eve', 'Maths': 88, 'English': 78, 'Science': 85}]

# create dataframe from list of dictionaries
df = pd.DataFrame(students)

# show dataframe
df.head()

avg_marks = df[['Maths', 'English', 'Science']].mean()

# create bar chart
plt.bar(avg_marks.index, avg_marks)

# set title and labels
plt.title('Average Marks by Subject')
plt.xlabel('Subject')
plt.ylabel('Average Marks')

# show plot
plt.show()
```
